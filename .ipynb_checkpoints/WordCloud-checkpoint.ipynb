```python
%pylab inline
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from collections import Counter
```

```python
punc = ',.!?-*&^%$#@'
text = "".join(open('R&J.txt', 'r').readlines())
```

```python
words = Counter([word.lower().strip(punc) for word in text.split() if len(word.strip(punc))>2])
```

```python
def plot(words):
    values = list(map(lambda x: x[1], words))
    labels = list(map(lambda x: x[0], words))
    plt.barh(range(len(words)), values, color='green')
    plt.yticks(range(len(words)), labels)
    plt.show
```

```python
plot(words.most_common()[0:15])
```

```python
wordsentence = " ".join(x[0] for x in words.most_common()[0:500] for times in range(0, int(x[1])))
```

```python
wordcloud = WordCloud(
                      font_path = '/Users/whitesky/Library/Fonts/CabinSketch-Bold.ttf',
                      stopwords=STOPWORDS,
                      background_color='white',
                      width=3000,
                      height=3000
                     ).generate(wordsentence)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
