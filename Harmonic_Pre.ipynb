```python
T= linspace(0,1, num=1000)
```

```python
plot(T)
```

```python
A=10 
ζ=0.1
ω=100
ϕ=0
```

```python
A, ζ, ω, ϕ
```

```python
Z=A*exp(-ζ*ω*T)* sin(sqrt(1-ζ**2) * ω*T + ϕ)
```

```python
plot (Z)
```

```python
def damping(A,ζ,ω, ϕ):
    Z=A*exp(-ζ*ω*T)* sin(sqrt(1-ζ**2) * ω*T + ϕ)
    plot(Z)
```

```python
from IPython.html.widgets import interact
interact(damping, A=(0,100), ζ=(0,1, 0.01), ω=(0, 1000), ϕ=(-10,10))
```
